```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Importamos las funciones desde modulo.py
from modulo import f, g

# Generación de puntos de entrada utilizando numpy
# Para f(x) = x^2 en el intervalo [-10, 10]
x_vals = np.linspace(-10, 10, 400)  # Generamos 400 puntos entre -10 y 10
y_vals_f = f(x_vals)  # Evaluamos f(x) = x^2

# Para g(x, y) = sin(x) * cos(y) en el dominio [-5, 5]
x_vals_g = np.linspace(-5, 5, 100)  # Generamos 100 puntos entre -5 y 5 para x
y_vals_g = np.linspace(-5, 5, 100)  # Generamos 100 puntos entre -5 y 5 para y
X_g, Y_g = np.meshgrid(x_vals_g, y_vals_g)  # Creamos una malla 2D para los valores de x e y
Z_g = g(X_g.flatten(), Y_g.flatten())  # Evaluamos g(x, y) = sin(x) * cos(y)
Z_g = np.array(Z_g).reshape(X_g.shape)  # Reshape para que Z_g sea una matriz 2D

# **Visualización 1: Gráfico 2D de f(x) = x^2**
plt.figure(figsize=(8, 6))
plt.plot(x_vals, y_vals_f, label=r"$f(x) = x^2$", color='b')
plt.title("Gráfico 2D de f(x) = x²")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()

# Guardamos el gráfico 2D como un archivo PNG
plt.savefig('grafico_f_x2.png', format='png')

# **Visualización 2: Gráfico 3D de g(x, y) = sin(x) * cos(y)**
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X_g, Y_g, Z_g, cmap='viridis', edgecolor='none')

ax.set_title(r"Gráfico 3D de $g(x, y) = \sin(x) \cdot \cos(y)$")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('g(x, y)')

# Guardamos el gráfico 3D como un archivo PNG
plt.savefig('grafico_g_xy_sin_cos.png', format='png')

# Mostrar los gráficos
plt.show()



```
