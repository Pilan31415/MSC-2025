```python
# Importa el módulo
from modulo import Punto2D

# Ejemplo 1: Crear el objeto Punto2D y mostrarlo
A = Punto2D(3, 4)
print(A)  # Debe imprimir (3,4)

# Ejemplo 2: Multiplicar el punto por un escalar
print(A(3))  # Debe imprimir (9, 12)

# Ejemplo 3: Sumar y restar puntos
B = Punto2D(5, 6)
C = 3 * A + B
print(f'C.x = {C.x}, C.y = {C.y}')  # Debe imprimir C.x = 14, C.y = 18

# Ejemplo 4: Graficar el punto
A.grafica()  # Debe mostrar la gráfica con el punto (3,4)

# Ejemplo 5: Calcular la distancia entre dos puntos y medir el tiempo
dist = A.distancia(B)  # Debe calcular la distancia y mostrar el tiempo de ejecución
print(f'Distancia entre A y B: {dist}')  # Debe imprimir la distancia entre A y B

# Ejemplo 6: Calcular la distancia al origen usando abs
print(abs(A))  # Debe imprimir la distancia de A al origen, que es 5
```
